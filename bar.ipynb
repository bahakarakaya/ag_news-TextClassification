```
import matplotlib.pyplot as plt


tests = [
    "1 - Basic Prompt",
    "2 - Role Assignment",
    "5 - One-Shot",
    "3 - Justification",
    "13 - Editor + One-Shot",
    "6 - Role + One-Shot",
    "14 - Archivist + One-Shot",
    "16 - Data Scientist + One-Shot",
    "15 - SME + One-Shot",
    "4 - Role + Justification",
    "7 - One-Shot + Justification",
    "8 - Role + One-Shot + Justification",
    "18 - Markdown + Multiple",
    "10 - Role + Multiple",
    "9 - Multiple Examples",
    "11 - Multiple + Justification",
    "12 - Role + Multiple + Justification"
]

efficiency = [
    0.855, 0.774, 0.568, 0.550, 0.506, 0.512, 0.494, 0.495, 0.477,
    0.503, 0.416, 0.360, 0.335, 0.244, 0.254, 0.194, 0.191
]

# Sort by efficiency and set figure size
tests_sorted, efficiency_sorted = zip(*sorted(zip(tests, efficiency), key=lambda x: x[1], reverse=True))

plt.figure(figsize=(12, 8))
bars = plt.barh(tests_sorted, efficiency_sorted, color='teal')
plt.xlabel("Accuracy / Token")
plt.title("Comparison of LLM Prompts")
plt.gca().invert_yaxis()

# Add value labels at the end of the bars
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.005, bar.get_y() + bar.get_height()/2, f"{width:.3f}", va='center')

plt.tight_layout()
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.show()
```
