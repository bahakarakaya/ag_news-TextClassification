```
import matplotlib.pyplot as plt


tests = [
    "1 - Temel İstem",
    "2 - Rol Atama",
    "5 - Tek Örnek",
    "3 - Gerekçe",
    "13 - Editor + Tek",
    "6 - Rol + Tek",
    "14 - Archivist + Tek",
    "16 - Data Scientist + Tek",
    "15 - SME + Tek",
    "4 - Rol + Gerekçe",
    "7 - Tek + Gerekçe",
    "8 - Rol + Tek + Gerekçe",
    "18 - Markdown + Çoklu",
    "10 - Rol + Çoklu",
    "9 - Çoklu Örnek",
    "11 - Çoklu + Gerekçe",
    "12 - Rol + Çoklu + Gerekçe"
]

verimlilik = [
    0.855, 0.774, 0.568, 0.550, 0.506, 0.512, 0.494, 0.495, 0.477,
    0.503, 0.416, 0.360, 0.335, 0.244, 0.254, 0.194, 0.191
]

# Grafik boyutu ve sıralama
tests_sorted, verimlilik_sorted = zip(*sorted(zip(tests, verimlilik), key=lambda x: x[1], reverse=True))

plt.figure(figsize=(12, 8))
bars = plt.barh(tests_sorted, verimlilik_sorted, color='teal')
plt.xlabel("Doğruluk / 100 Token")
plt.title("LLM İstemlerinin Verimlilik Karşılaştırması")
plt.gca().invert_yaxis()

# Barların sonuna değer etiketleri
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.005, bar.get_y() + bar.get_height()/2, f"{width:.3f}", va='center')

plt.tight_layout()
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.show()
```
